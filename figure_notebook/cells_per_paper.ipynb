```python
import polars
import plotly.express as px
tb = polars.read_csv("http://www.nxn.se/single-cell-studies/data.tsv", separator= "\t", ignore_errors = True)
```

```python
tb = tb.with_columns(year = tb["Date"].cast(str).apply(lambda s: s[0:4]).cast(int))
tb = tb.with_columns(
    n_cells = tb["Reported cells total"].apply(lambda s: int(s.replace(",", "")))
)

fig = px.box(tb.to_pandas(), x="year", y="n_cells")

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Number of cells",
    font=dict(
        size=18,
    )
    
)
# Limit yaxis
fig.update_yaxes(range=[0, 400000])
fig.show()
fig.write_image("number_of_cells.png")
```
